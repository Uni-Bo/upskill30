```python
# Load a new profile
profile = instaloader.Profile.from_username(L.context, '_basobdutta_')

# Get all posts in a generator object
posts = profile.get_posts()

# Iterate and download
for index, post in enumerate(posts, 1):
    L.download_post(post, target=f"{profile.username}_{index}")
```

```python
import matplotlib.pyplot as plt
posts = profile.get_posts()

# Initialize lists to store post likes and captions
likes = []
captions = []

# Iterate over the latest posts and extract likes and captions
for post in posts:
    likes.append(post.likes)
    captions.append(post.caption)

# Create a bar chart showing the number of likes for each post
plt.figure(figsize=(10, 6))
plt.bar(range(len(likes)), likes, color='skyblue')
plt.xlabel('Post Number')
plt.ylabel('Number of Likes')
plt.title('Number of Likes for Latest Instagram Posts')
plt.xticks(range(len(likes)), [f"Post {i+1}" for i in range(len(likes))], rotation=45)
plt.tight_layout()
plt.show()
```

```python
import instaloader
import matplotlib.pyplot as plt

# Get the latest posts
posts = profile.get_posts()

# Initialize counters for different post types
image_count = 0
video_count = 0
carousel_count = 0

# Iterate over the latest posts and categorize them based on type
for post in posts:
    if post.is_video:
        video_count += 1
    elif post.typename == 'GraphSidecar':
        carousel_count += 1
    else:
        image_count += 1

# Data for pie chart
labels = ['Images', 'Videos', 'Carousel']
sizes = [image_count, video_count, carousel_count]
colors = ['skyblue', 'lightcoral', 'lightgreen']

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Post Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```
